```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import datetime as dt
```

```python
def mean_mode_median_vendor_seating(df,seats_input):
    #activity_date column conversion from pd object to datetime64[ns] data type
    df['activity_date'] = pd.to_datetime(df['activity_date'])
    
    #Dropping the rows that have 'inactive' status in the program_status column
    df = df[df.program_status != "INACTIVE"]
    
    #User input for entering the desired seating number or type to project the mean,mode and median
    #for the inspection score per year
    seats_column = df.loc[df['seating_number_type'] == seats_input]
    
    #Creating a boolean series for each year
    year_2019 = seats_column['activity_date'].dt.year == 2019
    year_2018 = seats_column['activity_date'].dt.year == 2018
    year_2017 = seats_column['activity_date'].dt.year == 2017
    year_2016 = seats_column['activity_date'].dt.year == 2016
    
    #Each DataFrame object projects the rows corresponding to each year
    seats_column_2019 = seats_column.loc[year_2019]
    seats_column_2018 = seats_column.loc[year_2018]
    seats_column_2017 = seats_column.loc[year_2017]
    seats_column_2016 = seats_column.loc[year_2016]
    
    #Each series contains the score column for each year
    score_seats_column_2019 = seats_column_2019['score']
    score_seats_column_2018 = seats_column_2018['score']
    score_seats_column_2017 = seats_column_2017['score']
    score_seats_column_2016 = seats_column_2016['score']
    
    #Mean, mode and median values for each year's inspection score, per vendor's seating 
    mean_2019 = round(score_seats_column_2019.mean(), 3)
    mode_2019 = score_seats_column_2019.mode()
    median_2019 = score_seats_column_2019.median()
    
    #Mean, mode and median values for each year's inspection score, per vendor's seating
    mean_2018 = round(score_seats_column_2018.mean(), 3)
    mode_2018 = score_seats_column_2018.mode()
    median_2018 = score_seats_column_2018.median()
    
    #Mean, mode and median values for each year's inspection score, per vendor's seating 
    mean_2017 = round(score_seats_column_2017.mean(), 3)
    mode_2017 = score_seats_column_2017.mode()
    median_2017 = score_seats_column_2017.median()
    
    #Mean, mode and median values for each year's inspection score, per vendor's seating 
    mean_2016 = round(score_seats_column_2016.mean(), 3)
    mode_2016 = score_seats_column_2016.mode()
    median_2016 = score_seats_column_2016.median()
    
    if mode_2019.empty:
        mode_2019 = mode_2019.to_list()
        mode_2019.append(0)
    
    if mode_2018.empty:
        mode_2018 = mode_2018.to_list()
        mode_2018.append(0)
    
    if mode_2017.empty:
        mode_2017 = mode_2017.to_list()
        mode_2017.append(0)
    
    if mode_2016.empty:
        mode_2016 = mode_2016.to_list()
        mode_2016.append(0)
    
    data = [[mean_2019, mean_2018, mean_2017, mean_2016],
       [mode_2019[0], mode_2018[0], mode_2017[0], mode_2016[0]],
       [median_2019, median_2018, median_2017, median_2016]]
    
    n = 4
    mean = data[0]
    mode = data[1]
    median = data[2]

    st = np.arange(n)
    width = 0.10

    plt.bar(st, mean, width, label = 'Mean')
    plt.bar(st + width, mode, width, label = 'Mode')
    plt.bar(st + width + width, median, width, label = 'Median')

    plt.ylabel('Inspection Score')
    plt.title('Mean, mode and median for the inspection score per year for the selected seating type')

    plt.xticks(st + width, (2019, 2018, 2017, 2016))
    plt.legend(loc='best')
    plt.grid(b=True, which='major', linestyle='-')
    plt.minorticks_on()
    plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
    plt.show()
```
